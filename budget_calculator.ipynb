```python

class Category:

    def __init__(self, name):
        self.name = name
        self.ledger = []
        self.balance = 0

    def deposit(self, amount, description=""):
        self.ledger.append({"amount": amount , "description": description})
        self.balance += amount
    
    def withdraw(self, amount, description=""):
        if self.check_funds(amount) == False:
            return False
        else:
            amount = -amount
            self.ledger.append({"amount": amount , "description": description})
            self.balance += amount
            return True
        
    def get_balance(self):
        return self.balance
    
    def transfer(self, amount, category):
        if self.check_funds(amount) == False:
            return False
        else:
            self.withdraw(amount, "Transfer to " + category.name)
            category.deposit(amount, "Transfer from " + self.name)
            return True
    def check_funds(self,amount):
        if amount > self.balance:
            return False
        else:
            return True

    def __repr__(self):
        st = self.name.center(30,"*") + "\n"
        for i in self.ledger:
            st += f"{i['description'][:23]:23}" + f"{i['amount']:>7.2f}" + "\n"
        st += "Total: " + str(self.balance)
        return st

def create_spend_chart(categories):
    
    if len(categories) == 0:
        return "No categories found"
    elif len(categories) > 4:
        return "Too many categories"



    total = 0
    for i in categories:
        total += i.balance
    
    percent = {}
    for i in categories:
        percent[i.name] = round((i.balance / total) * 100, -1)
        

    st = "Percentage spent by category\n"

    for i in range(100, -10, -10):

        if i == 100:
            st += str(i) + "|" + " "
        elif i == 0:
            st += "  " + str(i) + "|" + " "
        else:
            st += " " + str(i) + "|" + " "


        if list(percent.values())[0] >= i:
            st += "o"
        else:
            st += " " 
        st += "  "
        for n in range(1, len(percent.values())):
            if list(percent.values())[n] >= i:
                st += "o"
            else:
                st += " "
            st += "  "
        st += "\n"
    
    st += "    " + "-" * (len(percent.values()) * 3 + 1) + "\n"

    max_len = 0
    for i in percent.keys():
        if len(i) > max_len:
            max_len = len(i)
    
    for j in range(max_len):
        st += "     "
        for i in range(len(percent.keys())):


            if j < len(list(percent.keys())[i]):
                st += list(percent.keys())[i][j]

            else:
                st += " "

            st += "  "
        st += "\n"


    return st






food = Category("Food")
food.deposit(500, "initial deposit")

shoes = Category("Shoes")
shoes.deposit(5000, "initial deposit")

books = Category("Books")
books.deposit(600, "initial deposit")

groceries = Category("Groceries")
groceries.deposit(300, "initial deposit")


my_list = [food, shoes, books]

a = create_spend_chart(my_list)

print (a)

print(food)
```
